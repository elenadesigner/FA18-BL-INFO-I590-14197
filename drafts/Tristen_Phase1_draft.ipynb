```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Read CSV
df = pd.read_csv("/Users/tristenrobrien/Desktop/Breast-Cancer-Wisconsin.csv")
print(df.head(10))
```

```python
#Replace ? by NaN in column A7. Use ____.replace('?', np.NaN) - specify A7 column.

print (df["A7"].replace('?', np.NaN))
```

```python
#Apply pandas function pd.to_numeric() for column A7

pd.to_numeric(df.A7, errors='coerce')
```

```python
#Report how many NaN using .isnull and .sum

how_many_nan = pd.isnull(df).sum()
print (how_many_nan)
```

```python
#Replace NaN values with the mean of column A7 with skipna=True
df.fillna(df.mean(skipna=True))
```

```python
#Provide the summary statistics - you can use describe() function
print (df.describe())
```

```python
#Find number of columns and number of rows - hint use ___.shape[]
print(str(df.shape[0]) + " rows")
print(str(df.shape[1]) + " columns")
```

```python
#Report how many unique id values (column Scn)

print (len(df["Scn"].unique()))
```

```python
#Draw histograms for columns A2-A10
df.loc[:, 'A2':'A10'].hist(bins = 7, color = 'green', alpha = 0.5, grid=False )
plt.show()
```
